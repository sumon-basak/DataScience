```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, MaxAbsScaler
```

```python
df_min_max = pd.read_csv('agora.csv')
df_min_max.isnull().sum()
```

```python
df_min_max.head()
```

# Min Max Scaler

```python
x = df_min_max.drop('Profit',axis=1)
y = df_min_max.Profit
min_max_scaler = MinMaxScaler(feature_range=(2,3))
df_min_max['Marketing Spend'] = min_max_scaler.fit_transform(x[['Marketing Spend']])
df_min_max['Administration'] = min_max_scaler.fit_transform(x[['Administration']])
df_min_max['Transport'] = min_max_scaler.fit_transform(x[['Transport']])

df_min_max
```

# Standard Scaler

```python
df_standard = pd.read_csv('agora.csv')

x = df_standard.drop('Profit',axis=1)
y = df_standard.Profit
standard_scaler = StandardScaler()
df_standard['Marketing Spend'] = standard_scaler.fit_transform(x[['Marketing Spend']])
df_standard['Administration'] = standard_scaler.fit_transform(x[['Administration']])
df_standard['Transport'] = standard_scaler.fit_transform(x[['Transport']])

df_standard
```

# Robust Scaler

```python
df_robust = pd.read_csv('agora.csv')

x = df_robust.drop('Profit',axis=1)
y = df_robust.Profit
robust_scaler = RobustScaler()
df_robust['Marketing Spend'] = robust_scaler.fit_transform(x[['Marketing Spend']])
df_robust['Administration'] = robust_scaler.fit_transform(x[['Administration']])
df_robust['Transport'] = robust_scaler.fit_transform(x[['Transport']])

df_robust
```
