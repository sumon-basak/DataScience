# One-Hot encoding

```python
df_one_hot = df.copy()


df_one_hot = pd.get_dummies(df_one_hot,columns=['Gender'] ,prefix = ['Gender'],drop_first=True)

df_one_hot
```

# Frequency Encoding

```python
data_freq = pd.read_csv('gender.csv')
fe = data_freq.groupby("Favorite Beverage").size()

fe_ = fe/len(data_freq)
data_freq["fe_Favorite_Beverage"] = data_freq["Favorite Beverage"].map(fe_).round(2)
data_freq.drop(['Favorite Beverage'],axis=1)
```
